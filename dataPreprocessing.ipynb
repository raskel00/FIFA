# prendre la derniere version de chaque joueur dans chaque fifa et les mettre dans de nouveux dataset

```python
v = [15, 16, 17, 18, 19, 20, 21, 22, 23]
taille_chunk = 1048576

for i in v:
    chemin_du_dataset = f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}_modifie.csv'

    # Utilisez 'chunksize' lors de la lecture pour diviser le dataset en chunks
    chunks = pd.read_csv(chemin_du_dataset, chunksize=taille_chunk)

    # Liste pour stocker les chunks modifiés
    chunks_modifies = []

    # Itérer sur les chunks pour convertir '4' et '3' en numériques
    for chunk in chunks:
        # Sélectionner la première occurrence de chaque joueur (supposant que les données sont triées)
        chunk = chunk.groupby('short_name').first().reset_index()

        # Ajouter le chunk modifié à la liste
        chunks_modifies.append(chunk)

    # Concaténer les chunks modifiés pour obtenir le dataset final
    dataset_modifie = pd.concat(chunks_modifies, ignore_index=True)
    dataset_modifie.sort_values(by='overall', ascending=False)

    # Afficher les premières lignes du dataset modifié
    print(dataset_modifie.head())

    # Enregistrer le dataset modifié dans un nouveau fichier CSV
    chemin_du_fichier_sortie = f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}_version_finaless.csv'
    dataset_modifie.to_csv(chemin_du_fichier_sortie, index=False)
    print(f"Le fichier '{chemin_du_fichier_sortie}' a été créé avec succès.")
```

## Prétraitement des Données FIFA 23 - Encodage et Stockage

*Description :*utilise LabelEncoder pour encoder certaines colonnes spécifiques des données des joueurs de FIFA. Les colonnes "preferred_foot", "league_name", "nationality_name", et "club_name" sont transformées à l'aide de LabelEncoder. Les données encodées sont ensuite stockées dans un dictionnaire, où chaque version du jeu FIFA  est représentée par un DataFrame modifié.

```python
from sklearn.preprocessing import LabelEncoder

v = [15, 16, 17, 18, 19, 20, 21, 22, 23]

# Créez un dictionnaire pour stocker les DataFrames de chaque version
data_dict = {}

for i in v:
    # Spécifiez le chemin vers le fichier CSV
    chemin_du_dataset = f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version2.csv'

    # Lire le fichier CSV dans un DataFrame
    data = pd.read_csv(chemin_du_dataset)

    # Initialisez LabelEncoder
    le = LabelEncoder()

    # Appliquez LabelEncoder à des colonnes spécifiques
    data["preferred_foot"] = le.fit_transform(data["preferred_foot"])

    # Stockez le DataFrame modifié dans le dictionnaire
    data_dict[f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version3.csv'] = data
```

# Encodage One-Hot des Positions des Joueurs

*Description :*utilisation de l'encodage One-Hot pour les positions des joueurs. Pour chaque version spécifiée dans la liste v, les données sont extraites du dictionnaire data_dict. Les positions multiples des joueurs sont transformées en listes et encodées en utilisant l'approche One-Hot. Le DataFrame résultant est sauvegardé dans un nouveau fichier CSV, facilitant l'utilisation ultérieure des données encodées dans des analyses ou des modèles d'apprentissage automatique.

```python
v = [15, 16, 17, 18, 19, 20, 21, 22, 23]

for i in v:
    # Accédez au DataFrame spécifique dans le dictionnaire
    data = data_dict[f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version3.csv']

    # Convertissez les listes en chaînes de caractères
    data['player_positions'] = data['player_positions'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

    # Divisez les positions multiples en listes
    data['player_positions'] = data['player_positions'].apply(lambda x: x.split(', '))

    # Appliquez One-Hot Encoding
    one_hot_encoded = data['player_positions'].apply(lambda x: pd.Series([1] * len(x), index=x)).fillna(0, downcast='int')

    # Concaténez les colonnes encodées avec le reste de la DataFrame
    data_encoded = pd.concat([data, one_hot_encoded], axis=1)

    # Sauvegardez le DataFrame modifié dans un nouveau fichier CSV
    data_encoded.to_csv(f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version3_encoded.csv', index=False)

    # Affichez le résultat si nécessaire
    print(data_encoded)
```

# Analyse des Valeurs et Ajustements des Évaluations des Joueurs FIFA 23

*Description :* Ce script Python parcourt différentes versions des données des joueurs FIFA  pour analyser les évaluations des joueurs et leurs ajustements. En utilisant des fichiers CSV spécifiques pour chaque version du jeu, le code divise les valeurs principales et les ajustements présents dans les colonnes telles que "ls", "st", "rs", etc. Ces valeurs sont sommées pour obtenir une nouvelle évaluation totale, qui est ensuite sauvegardée dans des fichiers CSV distincts. Ce processus facilite une analyse approfondie des performances des joueurs à travers les différentes versions du jeu FIFA.

```python
import re

v = [15, 16, 17, 18, 19, 20, 21, 22, 23]

for i in v:
    # Accédez au DataFrame spécifique dans le dictionnaire
    data = pd.read_csv(f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version3_encoded.csv')

    # Liste des colonnes à traiter
    columns_to_process = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm',
                           'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']

    # Définissez une fonction pour calculer la valeur totale
    def calculate_total(row):
        values = list(map(int, re.split('[-+]', row)))
        return sum(values)

    # Appliquez la fonction à toutes les colonnes spécifiées
    for col in columns_to_process:
        data[col] = data[col].apply(calculate_total)

    # Sauvegardez le DataFrame modifié dans un nouveau fichier CSV
    data.to_csv(f'/gdrive/MyDrive/FIFA-23-ML-Project/male_players_{i}version4.csv', index=False)

    print(data)
```
